```python
import pandas as pd
```

```python
df = pd.read_csv('movie_dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['key', 'MoviePoster'])
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset='URL')
```

```python
df['GenreList'] = df['Genres'].str.split(' / ')
# Get all unique genres across the entire dataset
all_genres = sorted(set(genre for sublist in df['GenreList'] for genre in sublist))
# Create a temporary DataFrame to store genre columns
genre_columns = pd.DataFrame()

# Populate the genre columns with 1s and 0s for each genre
for genre in all_genres:
    genre_columns[genre] = df['GenreList'].apply(lambda x: 1 if genre in x else 0)

# Append genre columns to the original DataFrame
df = pd.concat([df, genre_columns], axis=1)

# Drop the temporary 'GenreList' column from the DataFrame
df = df.drop(columns=['GenreList'])
```

```python
df['UploadedTime'] = pd.to_datetime(df['UploadedTime'])
```

```python
df.info()
```

```python
df = df.fillna(0)

df.notnull().sum()
```

```python
# Convert non-datetime values to NaT
df['UploadedTime'] = pd.to_datetime(df['UploadedTime'], errors='coerce')

# Sort the dataframe by 'UploadedTime'
df = df.sort_values(by='UploadedTime', ascending=True)
```

```python
df.insert(0, 'id', range(1, len(df) + 1))
```

```python
# Convert non-datetime values to NaT
df['UploadedTime'] = pd.to_datetime(df['UploadedTime'], errors='coerce')

# Sort the dataframe by 'UploadedTime'
df = df.sort_values(by='UploadedTime', ascending=False)
```
